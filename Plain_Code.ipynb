```python
import pandas as pd
import numpy as np

date_range = pd.date_range(start="2020-01-01", periods=3*365*24, freq="h")

np.random.seed(42)

df = pd.DataFrame({
    "timestamp": date_range,
    "energy_consumption": 200 + 20*np.sin(np.linspace(0, 50, len(date_range))) + np.random.normal(0, 5, len(date_range)),
    "temperature": 25 + 10*np.sin(np.linspace(0, 20, len(date_range))) + np.random.normal(0, 2, len(date_range)),
    "humidity": 60 + 15*np.sin(np.linspace(0, 15, len(date_range))) + np.random.normal(0, 3, len(date_range)),
    "solar_radiation": np.clip(100*np.sin(np.linspace(0, 80, len(date_range))) + np.random.normal(0, 10, len(date_range)), 0, None),
    "wind_speed": 5 + 2*np.sin(np.linspace(0, 30, len(date_range))) + np.random.normal(0, 1, len(date_range))
})

df.to_csv(r"C:/Users/hp/Desktop/Shalini - Project/Dataset.csv", index=False)
```
